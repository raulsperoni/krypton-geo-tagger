```python
# -*- coding: utf-8 -*- 
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix as cm
from lib import geoSearch
import json
import logging
%matplotlib inline  
from elasticsearch import Elasticsearch
es = Elasticsearch([{'host': 'elasticsearch-geo', 'port': 9200}])
from elasticsearch.client import IndicesClient
ic = IndicesClient(es)
GS = geoSearch.GeoSearch("elasticsearch-geo",9200)
geoSearch.logger.setLevel(logging.DEBUG)
from shapely.geometry import mapping, shape
```

```python
test_set = pd.read_csv('20.04.2020-ResultadosLocationEscucha.csv', index_col=0, encoding='utf-8')
```

```python
def analyzedText(text, analyzer='normal_analyzer'):
    results = ic.analyze(index="montevideo", body=
    {
      "text": text, 
        "analyzer": analyzer
    })
    return ' '.join([token['token'] for token in results['tokens']])
```

```python
geoSearch.logger.setLevel(logging.INFO)
for index, row in test_set[(test_set['score']>10) & (test_set['score']<=100)].iterrows():
    matches = GS.complete_search(row.text,50)
    print(index, row.text)
    for key, match in matches[:1]:
        print('✔️ MATCH: {:.3f}{: >20}{: >50}'.format(
            match['score'], 
            match['field'], 
            match['name']
        ))
    print('\n')
```
