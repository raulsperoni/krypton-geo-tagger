```python
import pandas as pd 
df = pd.read_csv("UY.txt",sep='\t',header=None) 
df.head()
```

```python
df = df[[0,1,2,3,4,5]]
```

```python
df.columns = ['id','name','name_accents','name_aliases','lat','lon']
df.head()
```

```python
min_lon, min_lat, max_lon, max_lat = [-56.4314,-34.938056,-56.022501,-34.701853]
```

```python
df_filtered = df[(df['lat'] >= min_lat) & (df['lat'] <= max_lat) & (df['lon'] >= min_lon) & (df['lon'] <= max_lon)]
```

```python
df_filtered.name_aliases.fillna(df.name, inplace=True)
```

```python
df_filtered.head()
```

```python
df_filtered.to_json('geonames_uy_montevideo.json',orient='records',lines=True)
```
